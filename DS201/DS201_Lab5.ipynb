```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
path ='/content/drive/MyDrive/Colab Notebooks/lab5.csv'
data = pd.read_csv(path)
```

```python
print(data.head())
print(data.shape)
```

```python
grdata = data.groupby(['State_name']).size()
print(grdata.head())
print(type(grdata))
```

```python
chart = grdata.plot(kind='bar', title = 'Number of universities per state', figsize= (15,8), fontsize = 10, color ='b', legend=False)
chart.set_xlabel('State', fontsize=15)
chart.set_ylabel ('Number of universities', fontsize = 15)
```
