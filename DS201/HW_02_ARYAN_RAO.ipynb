```python
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
```

```python
path ='/content/drive/MyDrive/Colab Notebooks/Lemonade2016.csv'
df = pd.read_csv(path)
```

```python
df.iloc[8,0] = '7/8/2016'
```

```python
df['Leaflets'].notnull()
```

```python
df.isna().sum()
```

```python
df['Leaflets'] = df['Leaflets'].fillna(round(df['Leaflets'].mean(),0))
```

```python
df.drop_duplicates()
```

```python
df = df.reset_index(drop=True)
```

```python
df['Sales'] = df['Lemon'] + df['Orange']
```

```python
df['Revenue'] = df['Sales'] * df['Price']
```

```python
df.head(50)
```

```python
plt.scatter(df['Leaflets'],df['Sales'])
plt.show()
```

```python
plt.hist(df['Revenue'])
plt.show()
```
