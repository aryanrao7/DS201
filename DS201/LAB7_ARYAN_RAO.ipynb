```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Downloading Data from website
!wget https://raw.githubusercontent.com/asukul/DS201/master/datasets/Automobile_price_data_Cleaned.csv
```

```python
#Preview first 5 rows
df = pd.read_csv('Automobile_price_data_Cleaned.csv')
df.head()
```

```python
df.describe()
```

```python
plt.hist(df['city-mpg'],bins=15)
plt.title('Histogram of automotive city-mpg')
plt.show()
```

```python
df1 = df.boxplot(column='city-mpg', by = 'make', rot=45, figsize = (15,8),fontsize= 12, grid = False)
df1.set_xlabel('city-mpg', fontsize=12)
df1.set_ylabel ('make', fontsize = 12)
plt.show()
```

```python
chart2 = df.plot.scatter(x = 'city-mpg', y = 'engine-size', c= 'curb-weight', fontsize = 12, sharex= False)
chart2.set_title('Scatterplot of engine-size versus city-mpg, grouped by curb-weight')
```
