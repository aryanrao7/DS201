```python
df = pd.read_csv('http://lidicky.name/DS201/Automobile_price_data_clean-f18.csv')
```

# **QUESTION 1**

```python
data = df.groupby(['make']).price.mean()
data.head(50)
```

```python
chart = data.plot(kind='bar', title = 'Average of Automobile price by make', figsize= (15,8), fontsize = 10, color ='b', legend=False)
chart.set_xlabel('Make', fontsize=15)
chart.set_ylabel ('Price', fontsize = 15)
```
